# Analyzing Records Of Cricket Players In Indian Premier League From 2008-2025

## Introduction

### 1. What is IPL ?
##### ->The Indian Premier League (IPL) is a professional Twenty20 (T20) cricket league in India, organised by the Board of Control for Cricket in India (BCCI).[1] Founded in 2007, it features ten city-based franchise teams.[2] The IPL is widely regarded as the world's most commercially successful cricket league. It is the 11th richest sporting league in the world by revenue.

### 2.Why to analyze records of IPL players.

#### -> The analysis of player records in IPL auctions is crucial for several reasons:
##### Performance Metrics: Teams analyze players' past performances, including batting and bowling averages, strike rates, economy rates, and match-winning abilities to identify trends and patterns. 
##### Market Value: Assessing the market value of players based on their performance, experience, and demand helps teams make informed decisions. 
##### Team Composition: Analyzing the existing squad composition to identify gaps and areas for improvement ensures teams build a balanced and effective squad. 
##### Auction Strategy: Developing data-driven auction strategies, including bidding patterns, player targeting, and budget allocation, allows teams to optimize their squads and maximize returns on investment. 

##### By leveraging data analytics, teams can make informed decisions, minimize risks, and maximize returns on investment, ultimately leading to a more successful IPL season.
##### Players analyze records before playing Dream11 to make informed decisions that can significantly impact their chances of success in the game. This analysis involves examining historical data, player performance metrics, and other relevant information to identify patterns and trends. By leveraging data-driven insights, players can make informed decisions about player selection, captaincy choices, and team composition, thereby increasing their chances of success.

## Data Cleaning

```python
import pandas as pd

print("Step 1.Load Data.")

df = pd.read_csv(
    'IPL.csv',
    usecols = [
        "date",
        "batting_team",
        "bowling_team",
        "player_of_match",
        "match_won_by",
        "venue",
        "year",
        "match_id"
    ]
                )

df.head(10)
```

```python
print("Step 2.Print Details Of Data.")
df.info()
```

```python
print("Step 3 : Check if data contains null data.")

print(df.isnull().sum())
```

```python
print("Step 4 : If data contains null rows then remove them.")

df = df.dropna()
```

```python
print("Step 5 : Check if data contains duplicate rows.")

print(df.duplicated().sum())
```

```python
print("Step 6 : If data contains duplicated rows then remove them.")

df = df.drop_duplicates()
```

```python
print("Step 7 : Standarize all formats of data to avoid further confusion")

df["batting_team"] = df["batting_team"].str.title()
df["bowling_team"] = df["bowling_team"].str.title()
df["player_of_match"] = df["player_of_match"].str.title()
df["match_won_by"] = df["match_won_by"].str.title()
df["venue"] = df["venue"].str.title()

df['date'] = pd.to_datetime(df['date'])
```

```python
print("Step 7 : After Cleaning print data again")

df.info()
df.head(10)
```

## ANALYSIS

```python
import matplotlib.pyplot as plt
```

```python
print("1.Number of matches happened per season : ")

n = df.groupby("year")["match_id"].nunique()

print(n)

n.plot(
    kind="bar"
)
plt.title("YEARS VS NUMBER OF MATCHES PLAYED")
plt.xlabel("YEAR")
plt.ylabel("MATCHES")
plt.show()
```

#### Insight : Number of matches kept fluctuating based on number of team participating

```python
print("2.MOST NUMBER OF WINS IN OVERALL SEASONS.")

n = df.groupby("match_won_by")["match_id"].nunique().sort_values(ascending=False).head(10)

print(n)

n.plot(
    kind = "pie",
    figsize = (5,5),
)

plt.title("MOST SUCCESSFULL TEAMS OF IPL IN TERMS OF MATCH WINS")
plt.show()
```

#### INSIGHT : WE FIND MUMBAI INDIANS HAS PLAYED MOST MATCHES FOLLOWED BY CHENNAI SUPER KINGS

```python
print("3.TOP PLAYER OF THE MATCHES")
n = df.groupby("player_of_match")["match_id"].nunique().sort_values(ascending=False).head(10)

print(n)

n.plot(
    kind="bar"
)
plt.ylabel("PLAYER OF THE MATCHES")
plt.title("TOP PLAYER OF THE MATCHES WINNERS")
plt.plot()
```

#### INSIGHT : TOP PERFORMERS ARE USUALLY FOREIGNERS EVEN THOUGH 7 OUT OF 11 ARE INDIANS

```python
df = pd.read_csv(
    "IPL.csv",
    usecols = [
        "batter",
        "runs_batter"
    ]
)
```

```python
print("4.TOP SCORERS OF THE LEAGUE")

n = df.groupby("batter")["runs_batter"].sum().sort_values(ascending=False).head(10)

print(n)

n.plot(
    kind = "bar",
)
plt.title("TOP SCORERS OF THE LEAGUE")
plt.ylabel("RUNS")
plt.xlabel("BATSMEN")
plt.show()
```

#### INSIGHT : BATTING IS MOSTLY DOMINATED BY INDIAN PLAYERS AS ONLY 1 FOREIGNER IS THERE IN TOP 5

```python
print("5.TOP 10 WICKET TAKERS")

df = pd.read_csv(
    "IPL.csv",
    usecols = [
        "bowler",
        "bowler_wicket"
    ]
)

n = df.groupby("bowler")["bowler_wicket"].sum().sort_values(ascending=False).head(10)
print(n)

n.plot(
    kind="bar",
)
plt.title("TOP 10 WICKET TAKERS")
plt.xlabel("BOWLER")
plt.ylabel("WICKETS")
plt.show()
```

#### INSIGHT : SPINNERS ARE DOMINATING THE TOP 10 BOWLERS SHOWING HIGH WICKET TAKING ABILITY.

```python
print("6.AVERAGE FIRST INNING SCORE VENUE WISE ")

df = pd.read_csv(
    "IPL.csv",
    usecols = [
        "venue",
        "runs_target"
    ]
)

#print(df.isnull().sum())
df = df.dropna()
#print(df.duplicated().sum())
df = df.drop_duplicates()

n = df.groupby("venue")["runs_target"].mean().sort_values(ascending=False).head(10)
print(n)

n.plot(
    kind="bar",
)
plt.title("TOP 10 BEST VENUES FOR BATTING FIRST")
plt.xlabel("VENUES")
plt.ylabel("AVERAGE FIRST INNING SCORE")
plt.show()
```

#### INSIGHT : STADIUMS OR GROUNDS SMALLER IN SIZE ARE EASIER TO SCORE ESPECIALLY IF THEY ARE LOCATED ON HIGHER ALTITUDE LIKE DHARMASHALA

```python
print("7.AVERAGE SECOND INNING SCORE VENUE WISE ")

df = pd.read_csv(
    "IPL.csv",
    usecols=[
        "match_id",
        "venue",
        "innings",
        "team_runs"
    ]
)

second = df[df["innings"] == 2]

matchscore = second.groupby(["match_id","venue"])["team_runs"].max().head(10)

avg = matchscore.groupby("venue").mean().sort_values(ascending=False).head(10)

avg.plot(
    kind = "bar"
)
plt.title("TOP 10 BEST VENUES FOR BATTING SECOND")
plt.ylabel("RUNS")
plt.show()
```

#### INSIGHT : IPL HISTORY SHOWS IT HAS BEEN ALWAYS DIFFICULT TO SCORE IN SECOND INNINGS.
